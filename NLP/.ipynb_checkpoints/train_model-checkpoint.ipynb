```python
import pandas as pd 

adata = pd.read_csv("./dialog_intent.csv", encoding='utf-8')
data_df = pd.DataFrame(adata)
data_df.head()
```

```python
from collections import Counter

intent = data_df["intent"]
cnt_intent = Counter(intent)
cnt_res = [(elem, cnt_intent[elem]) for elem in cnt_intent]
cnt_res.sort(key=lambda x:x[1], reverse=True)
len(cnt_res), cnt_res
```

```python
question_df = data_df[data_df["qa_type"] != "a"]

intent_dict = dict() 
cnt_dict = dict()
for i in range(len(question_df)): 
    row = question_df.iloc[i] 
    q, intent = row["question"], row["intent"]
    if intent not in intent_dict: 
        intent_dict[intent] = [q] 
        cnt_dict[intent] = 1 
    else: 
        if q not in intent_dict[intent]: 
            intent_dict[intent].append(q) 
            cnt_dict[intent] += 1
intent_dict
```

```python
que_intent = list() 
final_intent = [elem for elem in cnt_dict] 
for k, v in intent_dict.items(): 
    if k not in final_intent: continue
    for elem in v: que_intent.append([elem, k])
len(que_intent)
```

```python
dialog_data = list() 
ans_data = list() 
mean_length = 0 
max_length = 0 
length_list = list() 
for que, intent in que_intent: 
    pos_res = list() 
    que = list(que.replace(" ", ''))
    mean_length += len(que)
    max_length = max(max_length, len(que))
    length_list.append(len(que)) 
    dialog_data.append(" ".join(que))
    ans_data.append(intent) 
max_length, mean_length / len(que_intent), dialog_data[:10]
```

```python
import seaborn as sns 
sns.distplot(length_list)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing import sequence
MAX_NB_WORDS = 10000

tokenizer = Tokenizer(num_words=MAX_NB_WORDS, char_level=False)
tokenizer.fit_on_texts(dialog_data)
word_seq_train = tokenizer.texts_to_sequences(dialog_data)
word_index = tokenizer.word_index
print("dictionary size: ", len(word_index))
word_seq_train[:3]
```

```python
max_seq_len = 30
data_X = sequence.pad_sequences(word_seq_train, maxlen=max_seq_len)
data_X[:3]
```

```python
intent2idx, idx2intent = dict(), dict() 
for idx, intent in enumerate(set(ans_data)): 
    intent2idx[intent] = idx
    idx2intent[idx] = intent 
len(intent2idx)
```

```python
idx2intent
```

```python
import codecs 
from tqdm import tqdm
import numpy as np

embeddings_index = dict()
f = codecs.open('./fasttext/wiki.ko.vec', encoding='utf-8')
for line in tqdm(f):
    values = line.rstrip().rsplit(' ')
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embeddings_index[word] = coefs
f.close()
'found %s word vectors' % len(embeddings_index)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout, Flatten
from tensorflow.keras.layers import Embedding, Conv1D, MaxPooling1D, GlobalMaxPooling1D, LeakyReLU, BatchNormalization 
from tensorflow.keras import regularizers
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, Callback
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.utils import plot_model

import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
data_Y = [intent2idx[intent] for intent in ans_data]
data_Y = to_categorical(data_Y) 
num_classes = data_Y.shape[1]
data_Y.shape
```

```python
tokenizer = Tokenizer(num_words=MAX_NB_WORDS, char_level=False)
tokenizer.fit_on_texts(dialog_data)
word_seq_train = tokenizer.texts_to_sequences(dialog_data)
word_index = tokenizer.word_index
print("dictionary size: ", len(word_index))
word_seq_train[:3]
```

```python
data_X = sequence.pad_sequences(word_seq_train, maxlen=max_seq_len)
data_X[:3]
```

```python
print('preparing embedding matrix...')
words_not_found = []
embed_dim = 300 

nb_words = min(MAX_NB_WORDS, len(word_index)) + 1
embedding_matrix = np.zeros((nb_words, embed_dim))
for word, i in word_index.items():
    if i >= nb_words:
        continue
    embedding_vector = embeddings_index.get(word)
    if (embedding_vector is not None) and len(embedding_vector) > 0:
        # words not found in embedding index will be all-zeros.
        embedding_matrix[i] = embedding_vector
    else:
        words_not_found.append(word)
print('number of null word embeddings: %d' % np.sum(np.sum(embedding_matrix, axis=1) == 0))
print(embedding_matrix.shape)
```

```python
from sklearn.model_selection import train_test_split

train_X, test_X, train_Y, test_Y = train_test_split(data_X, data_Y, 
                                                   test_size=0.3, 
                                                   random_state=0, 
                                                   shuffle=True, 
                                                   stratify=data_Y
                                                   )
len(train_X), len(train_Y), len(test_X), len(test_Y)
```

```python
model = Sequential()
model.add(Embedding(nb_words, embed_dim,
                    weights=[embedding_matrix], 
                    input_length=max_seq_len, 
                    trainable=False))
model.add(Conv1D(32, 8, activation='relu', padding='same'))
model.add(MaxPooling1D(2))
model.add(Conv1D(16, 4, activation='relu', padding='same'))
model.add(GlobalMaxPooling1D())
model.add(Dense(32, 
                kernel_initializer='he_normal',
                kernel_regularizer=regularizers.l2(0.005)) )
model.add(BatchNormalization())
model.add(LeakyReLU(alpha=0.1))
model.add(Dropout(0.5))    
model.add(Dense(16, 
                kernel_initializer='he_normal',
                kernel_regularizer=regularizers.l2(0.005)) )
model.add(BatchNormalization())
model.add(LeakyReLU(alpha=0.1))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', 
              metrics=['accuracy'])
model.summary()
```

```python
plot_model(model, rankdir='LR', show_shapes=True)
```

```python
early_stopping_callback = EarlyStopping(monitor='val_accuracy',
                                        patience=100)

history = model.fit(train_X, train_Y, 
                    batch_size=64, 
                    epochs=1000,
                    validation_data=(test_X, test_Y),
                    callbacks=[early_stopping_callback],
                    verbose=2)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
def train_model(epochs, num): 
    model = Sequential()
    model.add(Embedding(nb_words, embed_dim,
                        weights=[embedding_matrix], 
                        input_length=max_seq_len, 
                        trainable=False))
    model.add(Conv1D(32, 8, activation='relu', padding='same'))
    model.add(MaxPooling1D(2))
    model.add(Conv1D(16, 4, activation='relu', padding='same'))
    model.add(GlobalMaxPooling1D())
    model.add(Dense(32, 
                    kernel_initializer='he_normal',
                    kernel_regularizer=regularizers.l2(0.005)) )
    model.add(BatchNormalization())
    model.add(LeakyReLU(alpha=0.1))
    model.add(Dropout(0.5))    
    model.add(Dense(16, 
                    kernel_initializer='he_normal',
                    kernel_regularizer=regularizers.l2(0.005)) )
    model.add(BatchNormalization())
    model.add(LeakyReLU(alpha=0.1))
    model.add(Dropout(0.5))
    model.add(Dense(num_classes, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', 
                  metrics=['accuracy'])
    history = model.fit(data_X, data_Y, 
                batch_size=64, 
                epochs=epochs,
                verbose=3)
    fin_acc = history.history['accuracy'][-1]
    model.save('./model_save/myModel_ver{}.h5'.format(num))
    return  fin_acc
```

```python
from time import time 
import datetime

start_time = time() 
total_acc = list()
epochs = 170
interval = 10
for i in range(1, 11):
    tmp_acc = train_model(epochs, i)
    total_acc.append((i, tmp_acc))
    epochs += interval
end_time = time() - start_time 
total_train_time = str(datetime.timedelta(seconds=end_time)).split(".")[0]
```
